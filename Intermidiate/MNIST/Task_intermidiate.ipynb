```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

Loading data 

```python
print("Downloading MNIST...")
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

Preprocessing the data Nueral Nets love numerical decimal values bewteen 0 to 1. so deividing by 255 as there are 256 total possibility for pixels

```python
train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255
```

Building model architecture

```python
model = tf.keras.models.Sequential([
    # The Eyes (Convolutional Layers)
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),

    # The Brain (Dense Layers)
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

Compiling the model using spare_categorical_crossentropy and adam optimizer for easy slope calculation

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Training the CNN

```python
print("\n--- Training the CNN ---")
history = model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_split=0.1)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"\n Test Accuracy: {test_acc*100:.2f}%")
```
